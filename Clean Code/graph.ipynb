```python
df_location = "df_migration_steps.csv"
df_geo_new_location = "df_geo_new.csv"
```

```python
df = pd.read_csv(df_geo_new_location)
```

```python
df
```

```python
df_cities = df.copy()
df_cities['code'] = df_cities['code'].astype(str).str[:4]
df_prefectures = df.copy()
df_prefectures['code'] = df_prefectures['code'].astype(str).str[:2]
```

```python
df_prefectures.columns
```

```python
df_cities
```

```python
#get the geodesic middle point between lon lat for df_cities with the same code use geopandas and use wgs84
import geopandas as gpd
from shapely.geometry import Point
    
geometry = [Point(xy) for xy in zip(df_cities['lon'], df_cities['lat'])]
gdf = gpd.GeoDataFrame(df_cities, geometry=geometry)
gdf.set_crs(epsg=4326, inplace=True)
gdf_grouped = gdf.dissolve(by='code', aggfunc='mean')
gdf_grouped = gdf_grouped.reset_index()
df_geo_new = pd.DataFrame(gdf_grouped.drop(columns='geometry'))
df_geo_new.to_csv('df_geo_cities.csv', index=False)
```

```python
geometry = [Point(xy) for xy in zip(df_prefectures['lon'], df_prefectures['lat'])]
gdf = gpd.GeoDataFrame(df_prefectures, geometry=geometry)
gdf.set_crs(epsg=4326, inplace=True)
gdf_grouped = gdf.dissolve(by='code', aggfunc='mean')
gdf_grouped = gdf_grouped.reset_index()
df_geo_new = pd.DataFrame(gdf_grouped.drop(columns='geometry'))
df_geo_new.to_csv('df_geo_prefectures.csv', index=False)
```
